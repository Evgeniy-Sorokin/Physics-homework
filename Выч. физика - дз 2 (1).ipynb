```python
def woodbury():
    print('Введите размер матрицы А (p):')
    p = int(input())
    print('Введите размер матрицы C (k):')
    k = int(input())
    a = np.random.random((p, p))
    a1 = np.linalg.inv(a)
    u = np.random.random((p, k))
    v = np.random.random((k, p))
    c = np.eye(k)
    t = np.linalg.inv(c + v @ a1 @ u)
    s = a1 - a1 @ u @ t @ v @ a1
    return s
```

```python
woodbury()
```

```python
def round_to_n(x, n):
    if x == 0:
        return x
    else:
        return round(x, -int(math.floor(math.log10(abs(x)))) + (n - 1))
```

```python
round_to_n(1.29343, 3)
```

```python
res = 0
for k in range (1, 3001):
    res = round_to_n(res + 1/k**2, 4)
print(res)
```

```python
new_res = 0
for k in range (3000, 0, -1):
    new_res = round_to_n(new_res + 1/k**2, 4)
print(new_res)
```
