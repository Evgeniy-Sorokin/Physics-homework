```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with np.load('/Users/20450/Downloads/data.npz') as data:
    A,C = data ['A'], data ['C']
```

```python
def mat2vec(A):
    h, w = A.shape
    a = np.zeros(h*w, dtype=A.dtype)
    A = np.flipud(A)
    for i, row in enumerate(A):
        a[i*w:i*w+w] = row
    return a

def vec2mat(a, shape):
    h, w = shape
    A = np.zeros(shape, dtype=a.dtype)
    for i in range(h):
        A[i,:] = a[i*w:i*w+w]
    return np.flipud(A)
```

```python
plt.imshow(A)
plt.show()
```

```python
a = mat2vec(A)
u, sig, v = np.linalg.svd(C, full_matrices=0)
w = np.diagflat(1/sig)
#w[:C.shape[1]-10,:C.shape[1]-10] = np.diag(sig)
a0 = v.T@w.T@u.T@a
A0 = vec2mat(a0, (16,51))
```
